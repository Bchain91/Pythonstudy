```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-4,4, 100)
y1 = 2**x
y2 = 3**x
y3 = 0.5**x

print('x = {}'.format(x))
print('y1 = {}'.format(y1))
print('y2 = {}'.format(y2))
print('y3 = {}'.format(y3))

plt.figure(figsize=(5,5))
plt.plot(x,y1, 'black', linewidth = 3, label = '$y=2^x$')
plt.plot(x,y2, 'cornflowerblue', linewidth = 3, label = '$y=3^x$')
plt.plot(x, y3, 'gray', linewidth = 3, label = '$y=0.5^x$')
plt.ylim(-2,6)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
plt.savefig('graph.png')
```

## 3차 함수 :f(x) = (x-2)*x*(x+2)

```python
def f(x):
    y = (x-2)*x*(x+2)
    return y

print('x = 1 ', f(1))
print('x = 2 ', f(2))
print('x = 3 ', f(3))

x = np.array([1,2,3]) #list -> np.ndarray
x = np.arange(-3,3.5, 0.5) #np.arange -> 범위 지정
x = np.linspace(-3,3, 100) #start, finish, samples
print('x [] = {}, type = {}'.format(x, type(x)))
print(f(x))


plt.figure(figsize = (5,5))
plt.plot(x, f(x))
plt.plot(x,f(x), 'black', linewidth = 3, label = '$y=(x-2)*x*(x+2)$')
plt.ylim(-30,30)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
#plt.savefig('graph.png')
```

```python
def f(x):
    y = (x-2)*x*(x+2)
    return y

def f2(x, w):
    y = (x-w)*x*(x+2)
    return y

x = np.linspace(-3,3, 50) #start, finish, samples
print('x [] = {}, type = {}'.format(x, type(x)))
print(f(x))


plt.figure(figsize = (5,5))
plt.plot(x, f(x))
plt.plot(x,f(x), 'blue', linewidth = 3, label = '$y=(x-2)*x*(x+2)$')

plt.plot(x,f2(x, w=0), 'black', linewidth = 3, label = '$w=0$')
plt.plot(x,f2(x, w=4), 'gray', linewidth = 3, label = '$y=(x-4)*x*(x+2)$')
plt.plot(x,f2(x, w=10), 'red', linewidth = 3, label = '$y=(x-10)*x*(x+2)$')

plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.ylim(-30,30)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
#plt.savefig('graph.png')
```

```python
def f3(x0, x1):
    r = 2*x0**2 + x1**2
    ans = r*np.exp(-r)
    return ans

#x0, x1에 따라서 각각 f3함수를 계산

xn = 50
x0 = np.linspace(-2,2, xn)
x1 = np.linspace(-2,2, xn)
y = np.zeros((len(x0), len(x1)))
print('x0 = {}'.format(x0))
print('x1 = {}'.format(x1))
print('y = {}'.format(y))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f3(x0[i0],x1[i1])

y = np.round(y, 1)
print('y = {}'.format(y))
plt.figure(figsize = (3.5,3))
plt.gray()
#plt.jet()
#plt.pink()
#plt.bone()

#plt.pcolor(y)
plt.contour(y)
plt.colorbar()
plt.show()
```

##  그래프 표현 - surface

```python
from mpl_toolkits.mplot3d import Axes3D

xn =50

x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)

xx0, xx1 = np.meshgrid(x0,x1)

plt.figure(figsize = (10,10))

for i in range(4):
    ax = plt.subplot(2,2,i+1, projection = '3d')
    ax.plot_surface(xx0,xx1, y, rstride = 1, cstride = 1, alpha = 0.3, 
                    color = 'blue', edgecolor='black')
    ax.set_zticks((0,0.2))
    if (i == 0):
        ax.view_init(0, 0)
    elif (i == 1):
        ax.view_init(30, -100)
    elif (i == 2):
        ax.view_init(40, -90)
    elif (i == 3):
        ax.view_init(30, -95)
```

## 그래프 표시 - contour

```python
xn =50
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((len(x0),len(x1)))

for row in range(xn):
    for col in range(xn):
        y[col, row] = f3(x0[row],x1[col])

xx0, xx1 = np.meshgrid(x0,x1)


plt.figure(figsize = (4,4))
cont = plt.contour(xx0, xx1, y, 5, colors = 'black')
cont.clabel(fmt='%3.2f', fontsize = 8)
plt.xlabel('$x_0$', fontsize = 14)
plt.ylabel('$x_1$', fontsize = 14)
```

## Quiz. 구구단 3단, 5단, 9단을 Plot(1개), SubPlot(4분할)로 그래프 표현

```python
def f(x, w):
    y = w*x
    return y

ytl=['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']

x = np.linspace(0,9, 10) #start, finish, samples

plt.figure(figsize = (5,5))
plt.plot(x, f(x, w=3), color = ytl[0], linewidth = 3, label = '$y=3*x$')
plt.plot(x, f(x, w=5), color = ytl[1], linewidth = 3, label = '$y=5*x$')
plt.plot(x, f(x, w=7), color = ytl[2], linewidth = 3, label = '$y=7*x$')
plt.plot(x, f(x, w=9), color = ytl[3], linewidth = 3, label = '$y=9*x$')
plt.title('$GuGu(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.ylim(0,90)
plt.xlim(3,10)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()

plt.figure(figsize = (5,5))
plt.subplots_adjust(wspace=0.5, hspace=0.5)
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.title(str(i*2+3) + 'Dan')
    plt.plot(x, f(x,i*2+3),color = ytl[i])
    plt.ylim(0,90)    
    plt.xlim(3,10)
    plt.grid(True)
plt.show()

#plt.savefig('graph.png')
```
